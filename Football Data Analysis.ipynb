```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('PL_21.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#country wise representation in PL
#asc. keyword used to print in desc. keyword
Nationality = df.groupby('Nation').size().sort_values(ascending = False)
Nationality.head(15).plot(kind= 'bar', figsize=(14,5),color = sns.color_palette("flare"))
```

```python
#clubs with biggest squad size

df['Team'].value_counts().nlargest(10).plot(kind = 'bar', color= sns.color_palette("Set2"))
```

```python
#clubs with smallest squad size
df['Team'].value_counts().nsmallest(5).plot(kind = 'bar', color= sns.color_palette("blend:#7AB,#EDA"))
```

```python
#players of different age group

Under20 = df[df['Age'] <=20]
Under25 = df[(df['Age'] >20) & (df['Age'] <=25)]
Under30 = df[(df['Age'] >25) & (df['Age'] <=30)]
Above30 = df[df['Age'] >30]

x = np.array([Under20['Player'].count(), Under25['Player'].count(), Under30['Player'].count(), Above30['Player'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", "30"]
plt.title('Players under each age group',fontsize = 20)
plt.pie(x, labels = mylabels, autopct="%.2f%%")
plt.legend()
plt.show()
```

```python
#total under 20 players in each club
U20_players = df[df['Age'] <=20]
U20_players['Team'].value_counts().plot(kind = 'bar', color = sns.color_palette("Spectral"))
```

```python
U20_players[U20_players["Team"] == 'Arsenal']
```

```python
#average squad age

plt.figure(figsize=(13,8))
sns.boxplot(x='Team', y='Age',data =df)
plt.xticks(rotation = 90)
```

```python
#average squad age by numbers
num_player = df.groupby('Team').size()
data = (df.groupby('Team')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
#total goals in the season

Total_Goals = df['Gls'].sum()
print(Total_Goals)
```

```python
#total assists in the season
Total_Assists = df['Ast'].sum()
print(Total_Assists)
```

```python
#total penalty goals in the season

Pen_Goals = df['PK'].sum()
print(Pen_Goals)
```

```python
#total penalties given

Total_Pens = df['PKatt'].sum()
print(Total_Pens)
```

```python
plt.figure(figsize= (13,6))
Pen_missed = Total_Pens - Pen_Goals
data = [ Pen_Goals , Pen_missed]
labels = ['Penalties scored' ,'Penalties missed']
color = sns.color_palette('muted')
plt.pie(data, labels=labels, colors= color,autopct= '%.1f%%')
plt.show()
```

```python
#different playing positions

df['Pos'].unique()
```

```python
#Best Goalscorers

Most_Goals = df[['Player', 'Team', 'Gls', 'MP']].nlargest(n=10, columns = 'Gls')
Most_Goals
```

```python
Club_goals = pd.DataFrame(df.groupby('Team', as_index = False)['Gls'].sum())
sns.set_theme(style="darkgrid",color_codes=True)
ax = sns.barplot(x='Team', y='Gls', data= Club_goals.sort_values(by="Gls"),palette='mako')
ax.set_xlabel("Team",fontsize=40)
ax.set_ylabel("Goals",fontsize=30)
plt.xticks(rotation=80)
plt.rcParams["figure.figsize"] = (20,20)
plt.title('Total Goals by each club',fontsize = 30)
```

```python
Most_goals_per90 = df[['Player', 'Team', 'Gp90', 'MP']].nlargest(n=10,columns = 'Gp90')
Most_goals_per90
```

```python
#total assists from each club

Club_assists = pd.DataFrame(df.groupby('Team', as_index = False)['Ast'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Team', y='Ast', data= Club_assists.sort_values(by="Ast"),palette='husl')
ax.set_xlabel("Team",fontsize=40)
ax.set_ylabel("Assists",fontsize=30)
plt.xticks(rotation=80)
plt.rcParams["figure.figsize"] = (10,5)
plt.title('Total Assists by each club',fontsize = 30)
```

```python
#Best playmakers

Most_assists = df[['Player', 'Team', 'Ast', 'MP']].nlargest(n=10, columns= 'Ast')
Most_assists
```

```python
Most_assists_per90 = df[['Player', 'Team', 'Ap90', 'MP']].nlargest(n=10,columns = 'Ap90')
Most_assists_per90
```

```python
#goals with and wothout assists

plt.figure(figsize=(13,6))
data = [Total_Goals - Total_Assists, Total_Assists]
labels = ['Goals without assists', 'Goals with assists']
color= sns.color_palette('hls')
plt.pie(data,labels = labels,colors = color,autopct='%.0f%%')
plt.show()
```

```python
yellow_cards = df.sort_values(by='CrdY', ascending =False)[:15]
plt.figure(figsize=(16,7))
plt.title("Players with most yellow cards")
c= sns.barplot(x= yellow_cards['Player'], y=yellow_cards['CrdY'], label = 'Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=60)
```

```python
red_cards = df.sort_values(by='CrdR', ascending =False)[:10]
plt.figure(figsize=(16,7))
plt.title("Players with most red cards")
c= sns.barplot(x= red_cards['Player'], y=red_cards['CrdR'], label = 'Players', color='red')
plt.ylabel('Number of Red Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=60)
```
